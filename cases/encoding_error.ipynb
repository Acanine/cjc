# Topic Modeling

```python
def jieba_tokenizer(doc):
    import jieba
    tokens = jieba.cut(doc)
    tokens = [el for el in tokens if len(el) > 1]
    str1 = open("/Users/chengjun/github/cjc/data/stopwords.txt", 'r').read()
    str1 = str1.split()
    str1 = [i.decode('utf8') for i in str1] # add this line
    tokens = [el for el in tokens if el not in str1]
    return tokens
```

```python
text_train_seg = []                 
for k, doc in enumerate(df1):
    if k % 100 == 0:
        print k
    text_train_seg.append(jieba_tokenizer(doc) )
```
